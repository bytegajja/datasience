## ANNALYSE DE SENTIMENT

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
# Création d'une instance de SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

data = df['Text']
    

data1 = pd.DataFrame(data)

def get_sentiment_score(Text):
    return analyzer.polarity_scores(Text)

# Appliquer la fonction à la colonne 'Text' du DataFrame avec tqdm pour afficher une barre de progression
data1['SentimentScores'] = df['Text'].apply(get_sentiment_score)

# Afficher le DataFrame avec les scores de sentiment ajoutés
print(data1)

```

```python


# Supposons que 'data1' est votre DataFrame contenant la colonne 'SentimentScores'

# Fonction pour extraire les parties négatives, positives et neutres des dictionnaires
def extract_sentiment_scores(row):
    return pd.Series([
        row['SentimentScores']['neg'],
        row['SentimentScores']['neu'],
        row['SentimentScores']['pos'],
        row['SentimentScores']['compound']
    ])

# Appliquer la fonction pour diviser la colonne en trois parties distinctes
data1[['Negative', 'Neutral', 'Positive', 'Compound']] = data1.apply(extract_sentiment_scores, axis=1)
```

```python
#supprime la ponctuation, normalise le texte en minuscules et élimine les mots vides (stopwords)
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Téléchargement des données NLTK (à faire une seule fois)
nltk.download('punkt')
nltk.download('stopwords')

# Fonction de prétraitement du texte
def preprocess_text(Text):
    # Suppression de la ponctuation
    Text = Text.translate(str.maketrans('', '', string.punctuation))

    # Mise en minuscules
    Text = Text.lower()

    # Tokenisation
    tokens = word_tokenize(Text)

    # Suppression des stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    return tokens

# Exemple d'utilisation de la fonction sur une colonne 'Text' du DataFrame
df['Processed_text'] = df['Text'].apply(preprocess_text)
print(df['Processed_text'])
```

```python
from collections import Counter

# Concaténation de tous les textes prétraités en une seule liste de mots
all_words = [word for sublist in df['Processed_text'] for word in sublist]

# Calcul de la fréquence des mots
word_freq = Counter(all_words)

# Affichage des 20 mots les plus fréquents
top_20_words = word_freq.most_common(20)
print("Les 20 mots les plus fréquents :\n", top_20_words)


```

```python
# Récupération des mots et de leur fréquence pour l'histogramme
top_words, freq = zip(*word_freq.most_common(20))

# Tracé de l'histogramme
plt.figure(figsize=(12, 6))
plt.bar(top_words, freq)
plt.xlabel('Mots')
plt.ylabel('Fréquence')
plt.title('Histogramme des 20 mots les plus fréquents')
plt.xticks(rotation=45)
plt.show()
```

## model 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import HashingVectorizer

# Utiliser les données après le prétraitement
text_data = [' '.join(tokens) for tokens in df['body_text_lemmatized']]
target = df['Sentiment']  # Supposons que vous avez cette colonne contenant les sentiments

# Division des données en ensembles de formation et de test
X_train, X_test, y_train, y_test = train_test_split(text_data, target, test_size=0.2, random_state=42)

# Initialisation du HashingVectorizer avec un nombre spécifié de features
hash_vectorizer = HashingVectorizer(n_features=1000, ngram_range=(1, 3))  # Vous pouvez ajuster n_features et ngram_range

# Vectorisation des données textuelles
X_train_hashed = hash_vectorizer.fit_transform(X_train)
X_test_hashed = hash_vectorizer.transform(X_test)

# Modèles à tester
models = {
    'Decision Tree': DecisionTreeClassifier(),
    'Random Forest': RandomForestClassifier(),
    
    
}

# Entraînement et évaluation des modèles
for model_name, model in models.items():
    print(f"Training {model_name}...")
    model.fit(X_train_hashed, y_train)
    
    predictions = model.predict(X_test_hashed)
    
    accuracy = accuracy_score(y_test, predictions)
    print(f"{model_name} Accuracy: {accuracy}")
    
    print(f"{model_name} Classification Report:")
    print(classification_report(y_test, predictions))
    print("=" * 50)
```
