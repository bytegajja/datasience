```python
#supprime la ponctuation, normalise le texte en minuscules et élimine les mots vides (stopwords)
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string

# Téléchargement des données NLTK (à faire une seule fois)
nltk.download('punkt')
nltk.download('stopwords')

# Fonction de prétraitement du texte
def preprocess_text(Text):
    # Suppression de la ponctuation
    Text = Text.translate(str.maketrans('', '', string.punctuation))

    # Mise en minuscules
    Text = Text.lower()

    # Tokenisation
    tokens = word_tokenize(Text)

    # Suppression des stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]

    return tokens

# Exemple d'utilisation de la fonction sur une colonne 'Text' du DataFrame
df['Processed_text'] = df['Text'].apply(preprocess_text)
print(df['Processed_text'])
```

```python
from collections import Counter

# Concaténation de tous les textes prétraités en une seule liste de mots
all_words = [word for sublist in df['Processed_text'] for word in sublist]

# Calcul de la fréquence des mots
word_freq = Counter(all_words)

# Affichage des 20 mots les plus fréquents
top_20_words = word_freq.most_common(20)
print("Les 20 mots les plus fréquents :\n", top_20_words)


```

```python
# Récupération des mots et de leur fréquence pour l'histogramme
top_words, freq = zip(*word_freq.most_common(20))

# Tracé de l'histogramme
plt.figure(figsize=(12, 6))
plt.bar(top_words, freq)
plt.xlabel('Mots')
plt.ylabel('Fréquence')
plt.title('Histogramme des 20 mots les plus fréquents')
plt.xticks(rotation=45)
plt.show()
```

## Vectorizing Data: N-Grams

```python
from sklearn.feature_extraction.text import HashingVectorizer
import numpy as np

# Supposons que df['body_text_lemmatized'] soit une liste de listes de tokens
# Convertir la liste de listes en une liste de chaînes de texte
text_data = [' '.join(tokens) for tokens in df['body_text_lemmatized']]

# Initialisation du HashingVectorizer avec un nombre spécifié de features
hash_vectorizer = HashingVectorizer(n_features=1000, ngram_range=(1, 1))  # Vous pouvez ajuster n_features

# Vectorisation des données textuelles
hashed_features = hash_vectorizer.transform(text_data)

# Conversion des features vectorisées en un tableau NumPy pour l'affichage
hashed_array = hashed_features.toarray()

# Affichage des features vectorisées
print("Matrice de features vectorisées :")
print(hashed_array)
print("\nShape de la matrice de features vectorisées :", hashed_array.shape)
```

# model vader et  DecisionTree

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
import pandas as pd

# Création d'une instance de SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

data = df['Text']

data1 = pd.DataFrame(data)

def get_sentiment_score(Text):
    return analyzer.polarity_scores(Text)

# Appliquer la fonction à la colonne 'Text' du DataFrame avec tqdm pour afficher une barre de progression
data1['SentimentScores'] = df['Text'].apply(get_sentiment_score)

# Ajouter la colonne 'Label' au DataFrame pour stocker les labels réels
data1['Label'] = df['Score']

# Convertir les scores de sentiment en labels prédits en utilisant une règle simple (par exemple, seuils de polarité)
data1['PredictedLabel'] = data1['SentimentScores'].apply(lambda x: 'positive' if x['compound'] >= 0.05 else ('negative' if x['compound'] <= -0.05 else 'neutral'))

# Calculer l'accuracy en comparant les labels prédits avec les labels réels
accuracy = (data1['PredictedLabel'] == data1['Label']).mean()

# Afficher le DataFrame avec les scores de sentiment ajoutés et l'accuracy
print(data1)
print(f'Accuracy: {accuracy:.2%}')
```

```python
# Fonction pour extraire les parties négatives, positives et neutres des dictionnaires
def extract_sentiment_scores(row):
    return pd.Series([
        row['SentimentScores']['neg'],
        row['SentimentScores']['neu'],
        row['SentimentScores']['pos'],
        row['SentimentScores']['compound']
    ])

# Appliquer la fonction pour diviser la colonne en trois parties distinctes
data1[['Negative', 'Neutral', 'Positive', 'Compound']] = data1.apply(extract_sentiment_scores, axis=1)
```

## model   DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import HashingVectorizer

# Utiliser les données après le prétraitement
text_data = [' '.join(tokens) for tokens in df['Processed_text']]
target = df['Score']  # Supposons que vous avez cette colonne contenant les sentiments

# Division des données en ensembles de formation et de test
X_train, X_test, y_train, y_test = train_test_split(text_data, target, test_size=0.2, random_state=42)

# Initialisation du HashingVectorizer avec un nombre spécifié de features
hash_vectorizer = HashingVectorizer(n_features=1000, ngram_range=(1, 3))  # Vous pouvez ajuster n_features et ngram_range

# Vectorisation des données textuelles
X_train_hashed = hash_vectorizer.fit_transform(X_train)
X_test_hashed = hash_vectorizer.transform(X_test)

# Modèles à tester
model = DecisionTreeClassifier()

# Entraînement et évaluation des modèles
print("Training Decision Tree Classifier...")
model.fit(X_train_hashed, y_train)

predictions = model.predict(X_test_hashed)

accuracy = accuracy_score(y_test, predictions)
print(f"Decision Tree Accuracy: {accuracy}")

print("Decision Tree Classification Report:")
print(classification_report(y_test, predictions))
print("=" * 50)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, precision_recall_curve, auc

# Supposons que 'y_test' sont les vraies étiquettes et 'predictions' sont les prédictions du modèle
# Assurez-vous que ces variables sont correctement définies dans votre code.

# Matrice de confusion
conf_matrix = confusion_matrix(y_test, predictions)

# Heatmap pour la matrice de confusion
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['1', '2', '3', '4', '5'], yticklabels=['1', '2', '3', '4', '5'])
plt.title('Matrice de Confusion')
plt.xlabel('Valeurs Prédites')
plt.ylabel('Valeurs Réelles')
plt.show()

# Courbes de précision et de rappel
precision, recall, _ = precision_recall_curve(y_test, predictions, pos_label=5)  # Assurez-vous de mettre la bonne étiquette positive (pos_label)
area_under_curve = auc(recall, precision)

plt.figure(figsize=(8, 6))
plt.step(recall, precision, color='b', alpha=0.2, where='post')
plt.fill_between(recall, precision, step='post', alpha=0.2, color='b')
plt.xlabel('Rappel (Recall)')
plt.ylabel('Précision (Precision)')
plt.title(f'Courbe de Précision-Rappel (AUC = {area_under_curve:.2f})')
plt.show()
```

#Courbe ROC (Receiver Operating Characteristic)
La courbe ROC est une représentation graphique de la performance d'un modèle de classification à différents seuils de classification. Elle peut être tracée en utilisant la bibliothèque scikit-learn.

```python
from sklearn.metrics import roc_curve, auc

fpr, tpr, _ = roc_curve(y_test, predictions, pos_label=5)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('Taux de Faux Positifs (FPR)')
plt.ylabel('Taux de Vrais Positifs (TPR)')
plt.title('Courbe ROC')
plt.legend(loc="lower right")
plt.show()
```
